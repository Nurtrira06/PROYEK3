```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breast-cancer.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
#ubah data kategorik  // memanipulasi data
df['diagnosis'].replace(['M', 'B'], ['Tumor Ganas (Terdeteksi Kanker)', 'Tumor Jinak (Tidak Terdeteksi Kanker)'], inplace=True)
df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']].corr(), annot=True, cmap='pink_r')
```

```python
# Bar Chart - Attribute diagnosis

plt.figure(figsize=(8,4))
sns.countplot(data = df,x = 'diagnosis')
```

```python
# pairplot

sns.pairplot(df, hue="diagnosis", vars=['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean'])
plt.show()
```

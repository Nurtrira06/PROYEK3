# MODELLING

```python

Diagnosis = pd.get_dummies(df['diagnosis'],drop_first=True)
Diagnosis.head()
```

```python
Diagnosis.rename(columns = {1 :'diagnosis'}, inplace = True)
Diagnosis.head()
```

```python
#ubah data kategorik  // memanipulasi data
Diagnosis['diagnosis'].replace([1, 0], 
                               ['Tumor Ganas (Terdeteksi Kanker)', 
                                'Tumor Jinak (Tidak Terdeteksi Kanker)'], 
                               inplace=True)
Diagnosis.head()
```

```python
df.drop(['diagnosis'],axis=1,inplace=True)
```

```python
df = pd.concat([df,Diagnosis],axis=1)
df.head()
```

```python
y = Diagnosis
y.head()
```

```python
X=df.drop(['diagnosis'],axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
persamaan = model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
print('\n')

print(confusion_matrix(y_test,pred))
```

```python
cm = confusion_matrix(y_test, pred)
sns.heatmap(cm, annot=True)
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import accuracy_score,
confusion_matrix,precision_score,f1_score,recall_score
print("accuracy score:", (accuracy_score(pred, y_test))*100, "%")
```
